# Markov Chain
从任意一个初始状态分布开始, 跟随着一个markov chain的状态转换矩阵, 可以最后收敛的细致平稳分布

```python
import numpy as np
# define transition matrix
P = np.array([0.9, 0.075, 0.025, 0.15, 0.8, 0.05, 0.25, 0.25, 0.5])
P = P.reshape(3,3) # P 是一个 3x3的转移矩阵
print(P)
```

```python
# initial states
e = np.array([0.5, 0.3, 0.2]) # 我们任意指定一个开始的分布

for i in range(40): #
    e = e.dot(P)
    print(f'interation i={i}, {e[0]:.3f} {e[1]:.3f} {e[2]:.3f}\n')
```

```python
# 我们换一个初始分布
# initial states
e = np.array([0.1, 0.1, 0.8])

for i in range(40): #
    e = e.dot(P)
    print(f'interation i={i}, {e[0]:.3f} {e[1]:.3f} {e[2]:.3f}\n')
```

我们发现，只要你的状态转移矩阵确定，那么无论从任何的初始状态分布出发，最后都会收敛到一个唯一的稳态状态分布
